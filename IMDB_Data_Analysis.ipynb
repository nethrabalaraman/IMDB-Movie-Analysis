```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/IMDB_Movies.csv")
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data["aspect_ratio"].fillna(int(data["aspect_ratio"].mean()),inplace=True)
```

```python
data["gross"].fillna(int(data["gross"].mean()),inplace=True)
```

```python
data["budget"].fillna(int(data["budget"].mean()),inplace=True)
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data=data.drop_duplicates()
```

```python
data.shape
```

```python
data['budget']=data['budget'].apply(lambda x: round(x/1000000,1))   
data['gross']=data['gross'].apply(lambda x: round(x/1000000,1))
data.head()
```

```python
data["profit"]=data["gross"]-data["budget"]
data.head()
```

```python
data=data.sort_values(by=['profit'],ascending=False)
data.head()
```

```python
import seaborn as sns
plt.figure(num=None, figsize=(5,5), dpi=80)   
plt.scatter(data['budget'], data['profit'])
plt.xlabel=("Budget")
plt.ylabel=("Profit") 
plt.show()
```

```python
plt.figure(num=None, figsize=(5,5), dpi=80)   
ploting=data[data.profit>100]
plt.scatter(ploting['budget'], ploting['profit'],edgecolors = 'b',facecolor='red') 
plt.xlabel=("Budget") 
plt.ylabel=("Profit") 
plt.title("Budget vs Profit")
plt.show()
```

```python
movie=data[data.profit>100]
top_5=movie[['movie_title']][0:5]
top_5
```

```python
IMDb_Top_250=data[['imdb_score','num_voted_users','movie_title','language']]
IMDb_sort= IMDb_Top_250.sort_values(by=['imdb_score'],ascending=False)
IMDb_Top_250=IMDb_sort[IMDb_Top_250.num_voted_users>25000]
IMDb_Top_250.head(250)
```

```python
IMDb_Top_250["rank"]=IMDb_Top_250['movie_title'].rank()
IMDb_Top_250['rank']=IMDb_Top_250['rank'].sort_values(ascending=True).values
IMDb_Top_250.head()
```

```python
df=data.groupby('director_name')
top_10_director=pd.DataFrame(df['imdb_score'].mean().sort_values(ascending=False)) 
top_10_director=top_10_director.head(10)
```

```python
top_10_director=top_10_director.sort_values(['imdb_score','director_name'],ascending=(False,True))
top_10_director
```

```python
data['genres']=data.genres.str.split('|')
data['genre_1']=data['genres'].apply(lambda x: x[0])  
data['genre_2']=data['genres'].apply(lambda x: x[1] if len(x)>1 else x[0])   #Extracting first and second
data.head()
```

```python
movie_segment = data.groupby(['genre_1','genre_2'])
movie_segment
```

```python
pop_genre=pd.DataFrame(movie_segment.gross.mean().sort_values(ascending=False) ) 
pop_genre[0:5]
```

```python
Meryl_Streep=data[['actor_1_name','movie_title','num_critic_for_reviews','num_user_for_reviews']]
Leo_Caprio=data[['actor_1_name','movie_title','num_critic_for_reviews','num_user_for_reviews']]
Brad_Pitt=data[['actor_1_name','movie_title','num_critic_for_reviews','num_user_for_reviews']]
```

Leo Caprio is the lead

Grouping into decade

```python
data['decade']=data['title_year'].apply(lambda x: (x//10) *10).astype(np.int64)  
data['decade']=data['decade'].astype(str)+'s'  
data=data.sort_values(['decade'])
data.head(5)
```

```python
by_decade=data.groupby('decade')
by_decade['num_voted_users'].sum()
by_decade=pd.DataFrame(by_decade['num_voted_users'].sum())
by_decade
```

```python
by_decade.plot.bar(figsize=(5,4),width=0.8,hatch="//",edgecolor='k')  
plt.xlabel=("Decade")
plt.ylabel=("Voted Users")
plt.title=("Voted users over Decades")
plt.yscale=('log')  
plt.show()
```
